# CREATE TRAINING AND TESTING DATASET

```python
admission_df.columns
```

```python
x = admission_df.drop(columns = 'Chance of Admit')
y = admission_df['Chance of Admit']
```

```python
print(x.shape)
x.head()
```

```python
print(y.shape)
y.head()
```

```python
x = np.array(x)
y = np.array(y).reshape(-1, 1)
y.shape
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler_x = StandardScaler()
x = scaler_x.fit_transform(x)
x
```

```python
# scaling the data before training the model
scaler_y = StandardScaler()
y = scaler_y.fit_transform(y)
y
```

```python
# spliting the data in to test and train sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
x_train
```

```python
y_train
```

# TRAIN AND EVALUATE A LINEAR REGRESSION MODEL

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, accuracy_score
```

```python
linreg = LinearRegression()
linreg.fit(x_train, y_train)
```

```python
accuracy_linreg = linreg.score(x_test, y_test)
accuracy_linreg
```

# TRAIN AND EVALUATE AN ARTIFICIAL NEURAL NETWORK

```python
import warnings
warnings.filterwarnings('ignore')
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.optimizers import Adam
```

```python
ANN_model = keras.Sequential()
ANN_model.add(Dense(50, input_dim = 7))
ANN_model.add(Activation('relu'))
ANN_model.add(Dense(150))
ANN_model.add(Activation('relu'))
ANN_model.add(Dropout(0.5))
ANN_model.add(Dense(150))
ANN_model.add(Activation('relu'))
ANN_model.add(Dropout(0.5))
ANN_model.add(Dense(50))
ANN_model.add(Activation('linear'))
ANN_model.add(Dense(1))
ANN_model.compile(loss = 'mse', optimizer = 'adam')
ANN_model.summary()
```

```python
ANN_model.compile(optimizer='Adam', loss='mean_squared_error')
```

```python
epochs_hist = ANN_model.fit(x_train, y_train, epochs = 100, batch_size = 20, validation_split = 0.2)
```

```python
result = ANN_model.evaluate(x_test, y_test)
accuracy_ANN = 1 - result
print("Accuracy : {}".format(accuracy_ANN))
```

```python
epochs_hist.history.keys()
```

```python
mp.figure(figsize = (10, 10))
_ = mp.plot(epochs_hist.history['loss'])
mp.title('Model Loss Progress During Training')
mp.xlabel('Epoch')
mp.ylabel('Training Loss')
_ = mp.legend(['Training Loss'])
```

# TRAIN AND EVALUATE A DECISION TREE AND RANDOM FOREST MODELS

```python
# Decision tree builds regression or classification models in the form of a tree structure. 
# Decision tree breaks down a dataset into smaller subsets while at the same time an associated decision tree is incrementally developed. 
# The final result is a tree with decision nodes and leaf nodes.
# Great resource: https://www.saedsayad.com/decision_tree_reg.htm
from sklearn.tree import DecisionTreeRegressor
DT = DecisionTreeRegressor()
DT.fit(x_train, y_train)
```

```python
accuracy_DT = DT.score(x_test, y_test)
accuracy_DT
```

```python
# Many decision Trees make up a random forest model which is an ensemble model. 
# Predictions made by each decision tree are averaged to get the prediction of random forest model.
# A random forest regressor fits a number of classifying decision trees on various sub-samples of the dataset and uses averaging to improve the predictive accuracy and control over-fitting. 
from sklearn.ensemble import RandomForestRegressor
RF = RandomForestRegressor(n_estimators = 200, max_depth = 10)
RF.fit(x_train, y_train)
```

```python
accuracy_RF = RF.score(x_test, y_test)
accuracy_RF
```

# CALCULATE REGRESSION MODEL KPIs

```python
y_pred_linreg = scaler_y.inverse_transform(linreg.predict(x_test))
y_pred_ANN = scaler_y.inverse_transform(ANN_model.predict_proba(x_test))
y_pred_DT = scaler_y.inverse_transform(DT.predict(x_test))
y_pred_RF = scaler_y.inverse_transform(RF.predict(x_test))
y_test_orig = scaler_y.inverse_transform(y_test)
```

The scatter plot shows us that almost all the models do a good job predicting the admission probablity since the data in the above charts is showing good amount of correlation.

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt

RMSE_linreg = float(format(np.sqrt(mean_squared_error(y_test_orig, y_pred_linreg)),'.3f'))
MSE_linreg = mean_squared_error(y_test_orig, y_pred_linreg)
MAE_linreg = mean_absolute_error(y_test_orig, y_pred_linreg)
r2_linreg = r2_score(y_test_orig, y_pred_linreg)
adj_r2_linreg = 1-(1-r2)*(n-1)/(n-k-1)

RMSE_ANN = float(format(np.sqrt(mean_squared_error(y_test_orig, y_pred_ANN)),'.3f'))
MSE_ANN = mean_squared_error(y_test_orig, y_pred_ANN)
MAE_ANN = mean_absolute_error(y_test_orig, y_pred_ANN)

RMSE_DT = float(format(np.sqrt(mean_squared_error(y_test_orig, y_pred_DT)),'.3f'))
MSE_DT = mean_squared_error(y_test_orig, y_pred_DT)
MAE_DT = mean_absolute_error(y_test_orig, y_pred_DT)

RMSE_RF = float(format(np.sqrt(mean_squared_error(y_test_orig, y_pred_RF)),'.3f'))
MSE_RF = mean_squared_error(y_test_orig, y_pred_RF)
MAE_RF = mean_absolute_error(y_test_orig, y_pred_RF)

print('\t\t\tLinear Regression\t\t\tANN\t\t\tDecision Tree\t\t\tRandom Forest\n')
print('RMSE\t\t\t{}\t\t\t\t\t{}\t\t\t{}\t\t\t\t{}\n'.format(RMSE_linreg, RMSE_ANN, RMSE_DT, RMSE_RF))
print('MSE\t\t\t{}\t\t\t{}\t{}\t\t\t{}\n'.format(MSE_linreg, MSE_ANN, MSE_DT, MSE_RF))
print('MAE\t\t\t{}\t\t\t{}\t{}\t\t{}\n'.format(MAE_linreg, MAE_ANN, MAE_DT, MAE_RF))
print('MSE\t\t\t{}\t\t\t{}\t{}\t\t\t{}\n'.format(MSE_linreg, MSE_ANN, MSE_DT, MSE_RF))
print('R^2\t\t\t{}\n'.format(r2_linreg))
print('R^2 Adjusted\t\t{}'.format(adj_r2_linreg))
```
